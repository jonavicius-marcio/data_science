## Métricas de Regressão

Os modelos de regressão são outra família de aprendizado de máquina e modelos estatísticos, usados para prever valores-alvo contínuos. 

Eles têm uma ampla gama de aplicações, desde previsão de preço da habitação, sistemas de preços de comércio eletrônico, previsão do tempo, previsão do mercado de ações, até super resolução de imagem, aprendizado de recursos via codificadores automáticos e compactação de imagem ou mesmo previsão de expectativa de vida.

Modelos como regressão linear, random forest, XGboost, rede neural convolucional, rede neural recorrente são alguns dos modelos de regressão mais populares.

As métricas usadas para avaliar esses modelos devem poder trabalhar em um conjunto de valores contínuos.

As principais métricas de regressão:

- MSE (Mean squared error)
- MAE (Mean absolute error)
- RMSE (Root mean squared error)
- R2 Score (Coeficiente de determinação)

```python
# Função para o plot da relação da variável alvo com alguns atributos
def get_pairs(data, alvo, atributos, n, fig_name = None):
    row_groups = [atributos[i:i+n] for i in range(0, len(atributos), n)]

    for ind in row_groups:
        plot = sns.pairplot(x_vars = ind, y_vars = alvo, data = data, kind = "reg", height = 3)
        if fig_name:
            plt.savefig(f"{fig_name}_{ind}.png")

    return
```

```python
# Função para o plot da matriz de correlação
def CorrMat(df, dropDuplicates=True, xrot=70, yrot=0, label='Variable'):

    # Excluir correlações duplicadas mascarando os valores superiores à direita
    if dropDuplicates:
        mask = np.zeros_like(df, dtype=np.bool)
        mask[np.triu_indices_from(mask)] = True

    # Definir cor do plano de fundo / estilo do gráfico
    sns.set_style(style='white')
    fig, ax = plt.subplots(figsize=(12, 10))

    # Adiciona mapa de cores do vermelho ao azul
    #cmap = sns.diverging_palette(250, 10, as_cmap=True)
    plt.title("Matriz de Correlação")

    # Desenha gráfico de correlação com ou sem duplicatas
    if dropDuplicates:
        sns.heatmap(df, mask=mask, square=True, linewidth=.5, cbar_kws={"shrink": .5}, ax=ax)
        plt.xlabel(label)
        plt.ylabel(label)
        plt.xticks(rotation=xrot)
        plt.yticks(rotation=yrot)

    else:
        sns.heatmap(df, square=True, linewidth=.5, cbar_kws={"shrink": .5}, ax=ax)
        plt.xlabel(label)
        plt.ylabel(label)
        plt.xticks(rotation=xrot)
        plt.yticks(rotation=yrot)
    return
```

### Modelagem Estatística

```python
# Primeira versão do modelo de regressão
formula = 'life_expectancy ~ bmi + hiv + gdp + thinness_till19 + inf_death + school + lifestyle + pop_size'
modelo_v1 = ols(formula, data = novo_vida_df_final).fit()
modelo_v1.summary()
```

```python
# Segunda versão do modelo de regressão sem variáveis que não são estatisticamente significantes
formula = 'life_expectancy ~ bmi + hiv + gdp + thinness_till19 + school + lifestyle'
modelo_v2 = ols(formula, data = novo_vida_df_final).fit()
modelo_v2.summary()
```

```python
# Aplicando escala aos dados
atributos = ['bmi', 'hiv', 'gdp', 'thinness_till19', 'school', 'lifestyle'] 

# Escala
scaler = StandardScaler()
scaler.fit(novo_vida_df_final[atributos])
scaled_features = scaler.transform(novo_vida_df_final[atributos])

# Criando um dataframe com dados em escala
scaled_features_df = pd.DataFrame(scaled_features, columns = atributos, index = novo_vida_df_final.index)
scaled_features_df.head()
```

```python
# Terceira versão do modelo
modelo_v3 = sm.OLS(endog = novo_vida_df_final['life_expectancy'], exog = sm.add_constant(scaled_features_df.values)).fit()
modelo_v3.summary()
```

## Modelagem Preditiva

```python
# Split dos dados em X e Y
X = novo_vida_df_final[['bmi', 'hiv', 'gdp', 'thinness_till19', 'inf_death', 'school', 'lifestyle']].values
y = novo_vida_df_final.life_expectancy.values.reshape(-1,1)
```

```python
# Split em treino e teste
x_treino, x_teste, y_treino, y_teste = train_test_split(X, y, test_size = 0.2, random_state = 0)
```

```python
# Modelo
modelo_lr_v1 = LinearRegression()
modelo_lr_v1.fit(x_treino, y_treino)
```

```python
# Previsões
y_predict = modelo_lr_v1.predict(x_teste)
```

```python
predict_df = pd.DataFrame({'Valor_Real': y_teste.flatten(), 'Valor_Previsto': y_predict.flatten()})
predict_df
```

```python
def create_scatter(x, y, title, xlabel, ylabel):
    fig, ax = plt.subplots(figsize=(8, 5))
    ax.scatter(x, y, color="blue", alpha=0.3)

    ax.set_title(title)
    ax.set_xlabel(xlabel)
    ax.set_ylabel(ylabel)

    ax.spines['top'].set_visible(False)
    ax.spines['right'].set_visible(False)
    return
```

```python
# Plot das previsões
create_scatter(predict_df.Valor_Real, predict_df.Valor_Previsto, 'Modelo', 'Previsões', 'Reais')
```

## Avaliação do Modelo

### Métricas

```python
# Print das métricas
print('Mean absolute error:', metrics.mean_absolute_error(y_teste, y_predict))
print('Mean squared error:', metrics.mean_squared_error(y_teste, y_predict))
print('Root mean squared error:', np.sqrt(metrics.mean_squared_error(y_teste, y_predict)))
print('R2 Score:', metrics.r2_score(y_teste, y_predict))
```

"Erro médio quadrático" (MSE) é talvez a métrica mais popular usada para problemas de regressão. Essencialmente, encontra o erro quadrado médio entre os valores previstos e reais.

Temos um modelo de regressão que prevê a expectativa de vida (definido como ŷᵢ) e temos ainda a expectativa de vida real (indicada com y). Então o MSE pode ser calculado como:

Às vezes, usamos o RMSE para ter uma métrica com escala como os valores-alvo, que é essencialmente a raiz quadrada do MSE.
